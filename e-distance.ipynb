```python
import numpy as np
import os
from PIL import Image
import helper
```

```python
# training data path
train_normal_path = "../Dataset/test/normal"
train_rosacea_path = "../Dataset/test/rosacea"
```

```python
# get the mean of the training data
sum = np.zeros((512,512,3))
count = 0
for filename in os.listdir(train_normal_path):
    file_path = os.path.join(train_normal_path, filename)
    img = Image.open(file_path)
    sum += np.array(img)
    count += 1
    # if count == 250:
    #     break
mean_normal = sum / count / 255


sum = np.zeros((512,512,3))
count = 0
for filename in os.listdir(train_rosacea_path):
    file_path = os.path.join(train_rosacea_path, filename)
    img = Image.open(file_path)
    sum += np.array(img)
    count += 1
    # if count == 250:
    #     break
mean_rosacea = sum / count / 255
```

```python
# Validation
TP, TN, FP, FN = 0, 0, 0, 0
val_normal_path = "../Dataset/val/normal"
count = 0
for filename in os.listdir(val_normal_path):
    count += 1
    file_path = os.path.join(val_normal_path, filename)
    sample = np.array(Image.open(file_path)) / 255
    d_n = helper.cal_e_distance(sample, mean_normal)
    d_r= helper.cal_e_distance(sample, mean_rosacea)
    if d_n < d_r:
        TN += 1
        print(f"d_n:{d_n} < d_r:{d_r}, {count}th in normal is true negative")

    else: # d_n > d_r
        FP += 1 
        print(f"d_n:{d_n} > d_r:{d_r}, {count}th in normal is false positive")
print(f"Validation on normal done, True Negative:{TN}, False Positive:{FP}\n")


count = 0
val_rosacea_path = "../Dataset/val/rosacea"
for filename in os.listdir(val_rosacea_path):
    count += 1
    file_path = os.path.join(val_rosacea_path, filename)
    sample = np.array(Image.open(file_path)) /255

    d_n = helper.cal_e_distance(sample, mean_normal)
    d_r= helper.cal_e_distance(sample, mean_rosacea)
    if d_n > d_r:
        TP += 1
        print(f"d_n:{d_n} > d_r:{d_r}, {count}th in rosacea is true positive")
    else: # d_n < d_r
        FN += 1
        print(f"d_n:{d_n} < d_r:{d_r}, {count}th in rosacea is false negative")
print(f"Validation on rosacea done, True Positive:{TP}, False Negative:{FN}")

print(TP,TN,FP,FN)
val_accuracy = (TP + TN) / (TP + TN + FP + FN)
print(f"TP:{TP}, TN:{TN}, FP:{FP}, FN:{FN}")
print(f"Val accuray:{val_accuracy}")




# Test
TP, TN, FP, FN = 0, 0, 0, 0
test_normal_path = "../Dataset/test/normal"
count = 0
for filename in os.listdir(test_normal_path):
    count += 1
    file_path = os.path.join(test_normal_path, filename)
    sample = np.array(Image.open(file_path)) / 255
    d_n = helper.cal_e_distance(sample, mean_normal)
    d_r= helper.cal_e_distance(sample, mean_rosacea)
    if d_n < d_r:
        TN += 1
        print(f"d_n:{d_n} < d_r:{d_r}, {count}th in normal is true negative")

    else: # d_n > d_r
        FP += 1 
        print(f"d_n:{d_n} > d_r:{d_r}, {count}th in normal is false positive")
print(f"Test on normal done, True Negative:{TN}, False Positive:{FP}\n")


count = 0
test_rosacea_path = "../Dataset/test/rosacea"
for filename in os.listdir(test_rosacea_path):
    count += 1
    file_path = os.path.join(test_rosacea_path, filename)
    sample = np.array(Image.open(file_path)) /255

    d_n = helper.cal_e_distance(sample, mean_normal)
    d_r= helper.cal_e_distance(sample, mean_rosacea)
    if d_n > d_r:
        TP += 1
        print(f"d_n:{d_n} > d_r:{d_r}, {count}th in rosacea is true positive")
    else: # d_n < d_r
        FN += 1
        print(f"d_n:{d_n} < d_r:{d_r}, {count}th in rosacea is false negative")
print(f"Test on rosacea done, True Positive:{TP}, False Negative:{FN}")

print(TP,TN,FP,FN)
test_accuracy = (TP + TN) / (TP + TN + FP + FN)
print(f"TP:{TP}, TN:{TN}, FP:{FP}, FN:{FN}")
print(f"Test accuray:{test_accuracy}")
```
